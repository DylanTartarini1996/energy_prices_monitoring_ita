```python
import pandas as pd
import yfinance as yf
```

# Natural Gas prices
Data here is collected through [Yahoo Finance](https://pypi.org/project/yfinance/) API

```python
symbol = 'TTF=F'
ticker = yf.Ticker(symbol)
```

```python
pd.Series(ticker.info).head(20)
```

```python
data = ticker.history(#period='20y',
                      interval='1wk',
                      start="2005-01-03",
                      end=None,
                      actions=True,
                      auto_adjust=True,
                      back_adjust=False)
data.info()
```

```python
data
```

```python
data.index = data.index.date
```

```python
data
```

```python
data = data[["Close"]]
data
```

```python
data.reset_index(inplace=True) 
```
